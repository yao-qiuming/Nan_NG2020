```python
filtered_proseq_df=pd.read_csv("filtered_geneproseqcount_combined_norm.csv",sep=",",header=0)
```

```python
filtered_proseq_df_2=filtered_proseq_df.loc[filtered_proseq_df['32hr_selected']==1]
gene_list=['HBG2', 'HBG1', 'HBB', 'HBD', 'HBBP1', 'HBE1', 'HBA1', 'HBA2', 'HBZ', 'HBM']
proseq_combined_norm=filtered_proseq_df
proseq_combined_norm_sub=proseq_combined_norm.loc[proseq_combined_norm['genename'].isin(gene_list)]


x=np.log2(np.add(proseq_combined_norm['norm_AAVS1_32hr_1'].tolist(),proseq_combined_norm['norm_AAVS1_32hr_2'].tolist()))
y=np.log2(np.add(proseq_combined_norm['norm_BCL11AKO_32hr_1'].tolist(),proseq_combined_norm['norm_BCL11AKO_32hr_1'].tolist()))
fig=plt.figure(figsize=(10,10))
plt.plot(x, y, 'o', color='grey',alpha=0.8,markersize=1)
plt.xlabel('log2 norm-Control-32h')
plt.ylabel('log2 norm-KO-32h')
#plt.plot([-4,4], [-4,4], 'b--',alpha=0.3)

for index in proseq_combined_norm_sub.index:
    x=np.log2(proseq_combined_norm_sub.loc[index]['norm_AAVS1_32hr_1']+proseq_combined_norm_sub.loc[index]['norm_AAVS1_32hr_2'])
    y=np.log2(proseq_combined_norm_sub.loc[index]['norm_BCL11AKO_32hr_1']+proseq_combined_norm_sub.loc[index]['norm_BCL11AKO_32hr_2'])
    plt.plot(x, y, 'o', color='red',alpha=0.9,markersize=1)
    t=proseq_combined_norm_sub.loc[index]['genename']
    plt.text(x+0.05, y+0.05, t, fontsize=12)

x1=np.log2(np.add(filtered_proseq_df_2['norm_AAVS1_32hr_1'].tolist(),filtered_proseq_df_2['norm_AAVS1_32hr_2'].tolist()))
y1=np.log2(np.add(filtered_proseq_df_2['norm_BCL11AKO_32hr_1'].tolist(),filtered_proseq_df_2['norm_BCL11AKO_32hr_1'].tolist()))
plt.plot(x1, y1, 'o', color='blue',alpha=0.8,markersize=1)

```
